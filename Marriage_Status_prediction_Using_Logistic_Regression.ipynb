# **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as pyplot
%matplotlib inline
```

# **Importing The Dataset**

```python
df = pd.read_csv('marital status.csv')
```

```python
df
```

# **Clecking Null Values**

```python
df.isnull().sum()
```

# **Standarization**

```python
handle = df['status'].median()
```

```python
handle
```

```python
df.status = df.status.fillna(handle)
```

```python
df
```

```python
df['status'].value_counts()
```

### **Separating  label and features(x=label , y=features)**

```python
x = df[['age']]
```

```python
x
```

```python
y = df['status']
```

```python
y
```

# Separating Test And Trainging **Data**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30,random_state=1)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

**Traing The Model**

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

**Accuracy **

```python
model.score(x_test,y_test)
```

Sigmoid Function Output


```python
model.predict_proba(x_test)
```
